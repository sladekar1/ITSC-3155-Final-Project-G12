```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import data
data=pd.read_csv("oscars_df.csv")

#Check what columns we have to decide wht we will use
pd.set_option("display.max.columns", None)
data.head(5)
```

```python
#some columns repit each other and some columns will not be useful for our app
# create copy of original dataset
data_new = data.copy()
# drop column we will not use
data_new.drop(["Unnamed: 0",'Oscar Year', 'IMDB Votes', "Genres",  "Authors", "Original Release Date", "Production Company", "Tomatometer Count",
              "Audience Count", "Tomatometer Top Critics Count", "Tomatometer Fresh Critics Count",  "Tomatometer Rotten Critics Count", "Streaming Release Date", "Audience Status"], axis=1, inplace= True)
data_new.head(5)
```

```python
data_new.info() 
```

```python
df1 = data_new[data_new.isna().any(axis=1)]
df1.head(5)
```

```python
data_new[["Directors", "Actors", "Tomatometer Status", "Content Rating"]]=data_new[["Directors", "Actors", "Tomatometer Status", "Content Rating"]].fillna("Unknown")
data_new[["Movie Info", "Critic Consensus"]]=data_new[["Movie Info", "Critic Consensus"]].fillna("No information")
```

```python
df1 = data_new[data_new.isna().any(axis=1)]
df1.head(5)
```

```python
data_new[['Tomatometer Rating','Audience Rating']] = data_new[['Tomatometer Rating','Audience Rating']].fillna(0.0)
```

```python
## check null values
data_new.isnull().sum().sort_values(ascending=False)[:10]
```

```python
# create new clean *.csv file
data_new.to_csv ("oscar_clean.csv")
```
